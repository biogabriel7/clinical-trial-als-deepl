```python
import re
from collections import Counter

def parse_multi_value_fields(df):
    """Parse and analyze multi-value fields in the dataset"""
    
    print("MULTI-VALUE FIELD ANALYSIS")
    print("=" * 60)
    
    # Define columns that likely contain multiple values
    multi_value_columns = [
        'Study Keywords',
        'MeSH Term', 
        'Primary Tested Drug: Therapeutic Class',
        'Other Tested Drug: Therapeutic Class',
        'Primary.Tested.Drug.Mechanism.Of.Action',
        'Other Tested Drug: Mechanism Of Action',
        'Oncology Biomarker',
        'Oncology Biomarker Common Use(s)',
        'Patient Age Group',
        'Trial Region',
        'Countries'
    ]
    
    # Analyze each multi-value column
    for col in multi_value_columns:
        if col in df.columns:
            print(f"\n{col}:")
            print("-" * 40)
            
            # Get non-null values
            non_null = df[col].dropna()
            if len(non_null) == 0:
                print("  No data available")
                continue
            
            # Check for common delimiters
            delimiters = [';', ',', '/', '|', ' and ', ' & ']
            delimiter_counts = {}
            
            for delimiter in delimiters:
                count = sum(1 for val in non_null if delimiter in str(val))
                if count > 0:
                    delimiter_counts[delimiter] = count
            
            if delimiter_counts:
                best_delimiter = max(delimiter_counts, key=delimiter_counts.get)
                print(f"  Primary delimiter: '{best_delimiter}' ({delimiter_counts[best_delimiter]} occurrences)")
                
                # Sample multi-value entries
                multi_value_samples = [val for val in non_null[:10] if best_delimiter in str(val)]
                if multi_value_samples:
                    print(f"  Sample multi-value entries:")
                    for i, sample in enumerate(multi_value_samples[:3]):
                        print(f"    {i+1}. {sample}")
            else:
                print("  No multi-value patterns detected")

def extract_keywords_features(df, config):
    """Extract individual keywords from Study Keywords column"""
    
    print(f"\nEXTRACTING STUDY KEYWORDS FEATURES")
    print("-" * 40)
    
    if 'Study Keywords' not in df.columns:
        return pd.DataFrame()
    
    # Extract all individual keywords
    all_keywords = []
    for keywords_str in df['Study Keywords'].dropna():
        # Split by semicolon and clean
        keywords = [kw.strip().lower() for kw in str(keywords_str).split(';')]
        all_keywords.extend(keywords)
    
    # Count keyword frequency
    keyword_counts = Counter(all_keywords)
    print(f"Total unique keywords: {len(keyword_counts)}")
    print(f"Most common keywords:")
    for keyword, count in keyword_counts.most_common(10):
        print(f"  {keyword}: {count}")
    
    # Create binary features for common keywords (appearing in at least 5 trials)
    common_keywords = [kw for kw, count in keyword_counts.items() if count >= 5]
    print(f"Keywords appearing in â‰¥5 trials: {len(common_keywords)}")
    
    # Create keyword features
    keyword_features = pd.DataFrame(index=df.index)
    
    for keyword in common_keywords:
        feature_name = f"keyword_{keyword.replace(' ', '_').replace('/', '_')}"
        keyword_features[feature_name] = df['Study Keywords'].apply(
            lambda x: 1 if pd.notna(x) and keyword in str(x).lower() else 0
        )
    
    return keyword_features

def extract_mesh_features(df, config):
    """Extract individual MeSH terms"""
    
    print(f"\nEXTRACTING MeSH TERM FEATURES")
    print("-" * 40)
    
    if 'MeSH Term' not in df.columns:
        return pd.DataFrame()
    
    # Extract all individual MeSH terms
    all_mesh = []
    for mesh_str in df['MeSH Term'].dropna():
        # Split by semicolon and clean
        terms = [term.strip() for term in str(mesh_str).split(';')]
        all_mesh.extend(terms)
    
    # Count term frequency
    mesh_counts = Counter(all_mesh)
    print(f"Total unique MeSH terms: {len(mesh_counts)}")
    print(f"Most common MeSH terms:")
    for term, count in mesh_counts.most_common(10):
        print(f"  {term}: {count}")
    
    # Create binary features for common terms (appearing in at least 10 trials)
    common_mesh = [term for term, count in mesh_counts.items() if count >= 10]
    print(f"MeSH terms appearing in â‰¥10 trials: {len(common_mesh)}")
    
    # Create MeSH features
    mesh_features = pd.DataFrame(index=df.index)
    
    for term in common_mesh:
        feature_name = f"mesh_{term.replace(' ', '_').replace(',', '').replace('(', '').replace(')', '').lower()}"
        mesh_features[feature_name] = df['MeSH Term'].apply(
            lambda x: 1 if pd.notna(x) and term in str(x) else 0
        )
    
    return mesh_features

def extract_drug_class_features(df, config):
    """Extract therapeutic class features"""
    
    print(f"\nEXTRACTING DRUG CLASS FEATURES")
    print("-" * 40)
    
    # Combine primary and other drug therapeutic classes
    all_classes = []
    
    for col in ['Primary Tested Drug: Therapeutic Class', 'Other Tested Drug: Therapeutic Class']:
        if col in df.columns:
            for class_str in df[col].dropna():
                # Split by semicolon and clean
                classes = [cls.strip() for cls in str(class_str).split(';')]
                all_classes.extend(classes)
    
    # Count class frequency
    class_counts = Counter(all_classes)
    print(f"Total unique therapeutic classes: {len(class_counts)}")
    print(f"Most common therapeutic classes:")
    for cls, count in class_counts.most_common(10):
        print(f"  {cls}: {count}")
    
    # Create binary features for common classes (appearing in at least 5 trials)
    common_classes = [cls for cls, count in class_counts.items() if count >= 5]
    print(f"Therapeutic classes appearing in â‰¥5 trials: {len(common_classes)}")
    
    # Create drug class features
    drug_features = pd.DataFrame(index=df.index)
    
    for cls in common_classes:
        feature_name = f"drug_class_{cls.replace(' ', '_').replace(',', '').replace('/', '_').lower()}"
        
        # Check both primary and other drug columns
        drug_features[feature_name] = 0
        for col in ['Primary Tested Drug: Therapeutic Class', 'Other Tested Drug: Therapeutic Class']:
            if col in df.columns:
                drug_features[feature_name] += df[col].apply(
                    lambda x: 1 if pd.notna(x) and cls in str(x) else 0
                )
        
        # Cap at 1 (binary feature)
        drug_features[feature_name] = (drug_features[feature_name] > 0).astype(int)
    
    return drug_features

def analyze_feature_importance_statistical(df, config):
    """Analyze features using statistical methods before neural network"""
    
    target_col = config['target_column']
    
    print(f"\nSTATISTICAL FEATURE ANALYSIS")
    print("=" * 60)
    
    # Simple categorical features (after we'll add the extracted multi-value features)
    simple_categorical = [
        'Trial Phase', 'Trial Status', 'Therapeutic Area', 'Patient Segment',
        'Sponsor Type', 'Patient Gender', 'Trial Region'
    ]
    
    from scipy.stats import chi2_contingency, fisher_exact
    from scipy.stats import mannwhitneyu
    
    significant_features = []
    
    print(f"Simple Categorical Feature Analysis:")
    print("-" * 40)
    
    for col in simple_categorical:
        if col in df.columns and df[col].nunique() < 20:  # Only analyze manageable categories
            try:
                # Create contingency table
                contingency = pd.crosstab(df[col], df[target_col])
                
                if contingency.shape == (2, 2):
                    # Use Fisher's exact test for 2x2 tables
                    _, p_value = fisher_exact(contingency)
                    test_name = "Fisher's"
                else:
                    # Use Chi-square for larger tables
                    _, p_value, _, _ = chi2_contingency(contingency)
                    test_name = "Chi-square"
                
                if p_value < 0.05:
                    significant_features.append((col, p_value, 'categorical'))
                    print(f"âœ“ {col}: p={p_value:.4f} ({test_name})")
                    
                    # Show the actual distribution
                    success_rates = df.groupby(col)[target_col].agg(['count', 'mean'])
                    print(f"  Success rates by category:")
                    for idx, row in success_rates.iterrows():
                        if row['count'] >= 5:  # Only show categories with enough samples
                            print(f"    {idx}: {row['mean']:.1%} ({row['count']} trials)")
                    print()
                    
            except Exception as e:
                print(f"âœ— {col}: Error - {e}")
    
    print(f"Found {len(significant_features)} statistically significant simple features")
    return significant_features

def preprocess_als_data_enhanced(df, config):
    """Enhanced preprocessing with proper multi-value field handling"""
    
    # First, analyze multi-value fields
    parse_multi_value_fields(df)
    
    # Extract features from multi-value fields
    keyword_features = extract_keywords_features(df, config)
    mesh_features = extract_mesh_features(df, config)
    drug_features = extract_drug_class_features(df, config)
    
    # Statistical analysis on simple features
    significant_features = analyze_feature_importance_statistical(df, config)
    
    # Exclude problematic columns
    exclude_cols = [
        config['target_column'], 
        config['id_column'], 
        'Protocol/Trial ID', 
        'Record URL', 
        'Trial Title',
        'Trial Objective',  
        'Study Design',
        'Treatment Plan',
        'Inclusion Criteria',
        'Exclusion Criteria',
        'Trial Results',
        'Trial Notes',
        'Supporting URLs',
        'Disposition of Patients',
        # Multi-value columns we're processing separately
        'Study Keywords',
        'MeSH Term',
        'Primary Tested Drug: Therapeutic Class',
        'Other Tested Drug: Therapeutic Class',
        'Oncology Biomarker',
        # Very sparse columns
        'Prior/Concurrent Therapy',
        'Decentralized (DCT) Attributes'
    ]
    
    # Select simple categorical and numerical features
    feature_cols = [col for col in df.columns if col not in exclude_cols]
    print(f"\nUsing {len(feature_cols)} simple feature columns")
    
    # Prepare base feature data
    X_base = df[feature_cols].copy()
    y = df[config['target_column']].copy()
    
    # Handle missing values in base features
    print(f"\nHandling missing values in base features:")
    for col in X_base.columns:
        missing_pct = X_base[col].isnull().mean()
        if missing_pct > 0.7:  # Remove features with >70% missing
            print(f"  Removing {col}: {missing_pct:.1%} missing")
            X_base = X_base.drop(columns=[col])
        elif X_base[col].dtype == 'object':
            X_base[col] = X_base[col].fillna('Missing')
        else:
            X_base[col] = X_base[col].fillna(X_base[col].median())
    
    # Smart categorical encoding for remaining features
    print(f"\nSmart categorical encoding:")
    for col in X_base.select_dtypes(include=['object']).columns:
        value_counts = X_base[col].value_counts()
        
        if len(value_counts) > 10:  # Too many categories
            # Keep only top 8 categories, group others as "Other"
            top_categories = value_counts.head(8).index
            X_base[col] = X_base[col].apply(lambda x: x if x in top_categories else 'Other')
            print(f"  {col}: {len(value_counts)} -> {X_base[col].nunique()} categories")
    
    # One-hot encode base categorical variables
    X_base_encoded = pd.get_dummies(X_base, drop_first=True)
    
    # Combine all features
    feature_sets = [X_base_encoded]
    feature_names = list(X_base_encoded.columns)
    
    if len(keyword_features.columns) > 0:
        feature_sets.append(keyword_features)
        feature_names.extend(keyword_features.columns)
        print(f"Added {len(keyword_features.columns)} keyword features")
    
    if len(mesh_features.columns) > 0:
        feature_sets.append(mesh_features)
        feature_names.extend(mesh_features.columns)
        print(f"Added {len(mesh_features.columns)} MeSH term features")
    
    if len(drug_features.columns) > 0:
        feature_sets.append(drug_features)
        feature_names.extend(drug_features.columns)
        print(f"Added {len(drug_features.columns)} drug class features")
    
    # Combine all feature sets
    X_combined = pd.concat(feature_sets, axis=1)
    print(f"Total features after combination: {X_combined.shape[1]}")
    
    # Feature selection: remove low-variance features
    from sklearn.feature_selection import VarianceThreshold
    
    variance_selector = VarianceThreshold(threshold=0.01)
    X_variance_filtered = variance_selector.fit_transform(X_combined.values)
    selected_features = [feature_names[i] for i in range(len(feature_names)) 
                        if i < len(variance_selector.get_support()) and variance_selector.get_support()[i]]
    
    print(f"After variance filtering: {len(selected_features)} features")
    
    # Convert to numpy and normalize
    X_array = X_variance_filtered.astype(np.float32)
    y_array = y.values.astype(np.float32)
    
    # Standardize features
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_array)
    
    # Create phase groups for stratification
    phase_groups = df['Trial Phase'].values
    
    return X_scaled, y_array, phase_groups, selected_features, scaler

def remove_phase_leakage(X_scaled, feature_names, config):
    """Remove trial phase features that leak target information"""
    
    if not config.get('exclude_phase_features', False):
        return X_scaled, feature_names
    
    print(f"\nREMOVING PHASE LEAKAGE FEATURES")
    print("-" * 40)
    
    # Identify phase-related features
    phase_features = [i for i, name in enumerate(feature_names) 
                     if 'trial_phase' in name.lower() or 'phase' in name.lower()]
    
    if phase_features:
        print(f"Removing {len(phase_features)} phase-related features:")
        for i in phase_features:
            print(f"  - {feature_names[i]}")
        
        # Remove phase features
        mask = [i for i in range(len(feature_names)) if i not in phase_features]
        X_filtered = X_scaled[:, mask]
        feature_names_filtered = [feature_names[i] for i in mask]
        
        print(f"Features after phase removal: {X_filtered.shape[1]}")
        return X_filtered, feature_names_filtered
    else:
        print("No phase features found to remove")
        return X_scaled, feature_names

# Enhanced preprocessing
print("Enhanced data preprocessing with multi-value field handling...")
X, y, phase_groups, feature_names, scaler = preprocess_als_data_enhanced(df, config)
print(f"\nFinal dataset shape: X={X.shape}, y={y.shape}")
print(f"Class distribution: {np.bincount(y.astype(int))} ({y.mean():.1%} positive)")
print(f"Total features extracted: {len(feature_names)}")
# Apply phase removal
X, feature_names = remove_phase_leakage(X, feature_names, config)
print(f"\nFinal dataset shape after phase removal: X={X.shape}")
```

```python
class LRP_Linear(nn.Module):
    def __init__(self, inp, outp, gamma=0.01, eps=1e-5):
        super(LRP_Linear, self).__init__()
        self.A_dict = {}
        self.linear = nn.Linear(inp, outp)
        nn.init.xavier_uniform_(self.linear.weight, gain=nn.init.calculate_gain('relu'))
        self.gamma = tc.tensor(gamma)
        self.eps = tc.tensor(eps)
        self.iteration = None

    def forward(self, x):
        if not self.training:
            self.A_dict[self.iteration] = x.clone()
        return self.linear(x)

    def relprop(self, R):
        device = next(self.parameters()).device
        A = self.A_dict[self.iteration].clone()
        A, self.eps = A.to(device), self.eps.to(device)

        Ap = A.clamp(min=0).detach().data.requires_grad_(True)
        Am = A.clamp(max=0).detach().data.requires_grad_(True)

        zpp = self.newlayer(1).forward(Ap)  
        zmm = self.newlayer(-1, no_bias=True).forward(Am) 
        zmp = self.newlayer(1, no_bias=True).forward(Am) 
        zpm = self.newlayer(-1).forward(Ap) 

        with tc.no_grad():
            Y = self.forward(A).data

        sp = ((Y > 0).float() * R / (zpp + zmm + self.eps * ((zpp + zmm == 0).float() + tc.sign(zpp + zmm)))).data
        sm = ((Y < 0).float() * R / (zmp + zpm + self.eps * ((zmp + zpm == 0).float() + tc.sign(zmp + zpm)))).data

        (zpp * sp).sum().backward()
        cpp = Ap.grad
        Ap.grad = None
        Ap.requires_grad_(True)

        (zpm * sm).sum().backward()
        cpm = Ap.grad
        Ap.grad = None
        Ap.requires_grad_(True)

        (zmp * sm).sum().backward()
        cmp = Am.grad
        Am.grad = None
        Am.requires_grad_(True)

        (zmm * sp).sum().backward()
        cmm = Am.grad
        Am.grad = None
        Am.requires_grad_(True)

        R_1 = (Ap * cpp).data
        R_2 = (Ap * cpm).data
        R_3 = (Am * cmp).data
        R_4 = (Am * cmm).data

        return R_1 + R_2 + R_3 + R_4

    def newlayer(self, sign, no_bias=False):
        if sign == 1:
            rho = lambda p: p + self.gamma * p.clamp(min=0)
        else:
            rho = lambda p: p + self.gamma * p.clamp(max=0)

        layer_new = copy.deepcopy(self.linear)
        try:
            layer_new.weight = nn.Parameter(rho(self.linear.weight))
        except AttributeError:
            pass

        try:
            layer_new.bias = nn.Parameter(self.linear.bias * 0 if no_bias else rho(self.linear.bias))
        except AttributeError:
            pass

        return layer_new

class LRP_ReLU(nn.Module):
    def __init__(self):
        super(LRP_ReLU, self).__init__()
        self.relu = nn.ReLU()

    def forward(self, x):
        return self.relu(x)

    def relprop(self, R):
        return R

class LRP_DropOut(nn.Module):
    def __init__(self, p):
        super(LRP_DropOut, self).__init__()
        self.dropout = nn.Dropout(p)

    def forward(self, x):
        return self.dropout(x)

    def relprop(self, R):
        return R

class ALS_Model(nn.Module):
    def __init__(self, n_features, config):
        super(ALS_Model, self).__init__()
        self.classname = 'ALS Clinical Trials Model'
        
        inp = n_features
        # Adjusted for 54 features - slightly larger network
        hidden = max(32, min(128, n_features * config['factor_hidden_nodes']))
        outp = 1
        
        # Build network layers dynamically based on config
        layers = []
        
        # Input layer
        layers.extend([
            LRP_DropOut(p=config['input_dropout']), 
            LRP_Linear(inp, hidden, gamma=config['lrp_gamma']), 
            LRP_ReLU()
        ])
        
        # Add hidden layers if specified
        for i in range(config['hidden_depth_simple']):
            layers.extend([
                LRP_DropOut(p=config['dropout']),
                LRP_Linear(hidden, hidden, gamma=config['lrp_gamma']),
                LRP_ReLU()
            ])
        
        # Output layer
        layers.extend([
            LRP_DropOut(p=config['dropout']),
            LRP_Linear(hidden, outp, gamma=config['lrp_gamma'])
        ])
        
        self.layers = nn.Sequential(*layers)
        
        # Count parameters
        nparams = sum(p.numel() for p in self.parameters() if p.requires_grad)
        print(f'{self.classname} contains {nparams:,} trainable parameters')
        
        # Show architecture
        if config['hidden_depth_simple'] == 0:
            print(f'Network architecture: {inp} -> {hidden} -> {outp}')
        else:
            hidden_layers = " -> ".join([str(hidden)] * config['hidden_depth_simple'])
            print(f'Network architecture: {inp} -> {hidden} -> {hidden_layers} -> {outp}')
        
        # Store config for regularization
        self.config = config

    def forward(self, x):
        return self.layers.forward(x)

    def relprop(self, R):
        assert not self.training, 'relprop does not work during training time'
        for module in self.layers[::-1]:
            R = module.relprop(R)
        return R
    
    def l1_regularization(self):
        """Compute L1 regularization term"""
        l1_reg = tc.tensor(0., device=next(self.parameters()).device)
        for param in self.parameters():
            l1_reg += tc.norm(param, 1)
        return l1_reg

print("Neural network classes defined âœ“")
```

```python
def train_model(X_train, y_train, X_val, y_val, model, config):
    """Train the ALS model with moderate class balancing"""
    
    # Clear GPU cache at start
    if tc.cuda.is_available():
        tc.cuda.empty_cache()
    
    # Convert to tensors
    X_train_tensor = tc.tensor(X_train, dtype=tc.float32, device=config['device'])
    X_val_tensor = tc.tensor(X_val, dtype=tc.float32, device=config['device'])
    train_targets = tc.tensor(y_train, dtype=tc.float32, device=config['device'])
    val_targets = tc.tensor(y_val, dtype=tc.float32, device=config['device'])
    
    # MODERATE CLASS BALANCING: Use fixed ratio instead of inverse frequency
    pos_count = y_train.sum()
    neg_count = len(y_train) - pos_count
    
    if config.get('use_class_weights', False):
        # Use moderate fixed weighting instead of inverse frequency
        pos_weight = tc.tensor([config.get('pos_weight_factor', 2.5)], device=config['device'])
        print(f"Using moderate class weights - pos_weight: {pos_weight.item():.2f}")
        print(f"(Instead of aggressive inverse frequency: {neg_count/pos_count:.2f})")
        criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
    else:
        criterion = nn.BCEWithLogitsLoss()
    
    optimizer = tc.optim.Adam(model.parameters(), lr=config['lr'], weight_decay=config['weight_decay'])
    scheduler = tc.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)
    
    best_val_f1 = 0.0  # Track best F1 score instead of loss
    patience_counter = 0
    training_history = []
    
    print(f"Training with moderate class balancing...")
    print(f"Train samples: {len(X_train)} (pos: {pos_count:.0f}, neg: {neg_count:.0f})")
    print(f"Class ratio: {pos_count/len(y_train):.3f} positive")
    
    # Training loop
    total_epochs = sum(config['reduce_lr_epochs'])
    for epoch in range(total_epochs):
        model.train()
        
        # Training step
        optimizer.zero_grad()
        train_outputs = model(X_train_tensor).squeeze()
        train_loss = criterion(train_outputs, train_targets)
        
        # Add L1 regularization if specified
        if 'l1_lambda' in config and config['l1_lambda'] > 0:
            train_loss += config['l1_lambda'] * model.l1_regularization()
        
        train_loss.backward()
        tc.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        
        # Validation step
        model.eval()
        with tc.no_grad():
            val_outputs = model(X_val_tensor).squeeze()
            val_loss = criterion(val_outputs, val_targets)
            
            # Calculate metrics with multiple thresholds
            train_probs = tc.sigmoid(train_outputs).cpu().numpy()
            val_probs = tc.sigmoid(val_outputs).cpu().numpy()
            
            # Find optimal threshold based on F1 score
            from sklearn.metrics import f1_score
            thresholds = np.arange(0.2, 0.8, 0.05)  # More reasonable threshold range
            best_f1 = 0
            best_thresh = 0.5
            
            for thresh in thresholds:
                f1 = f1_score(y_val, (val_probs > thresh).astype(int), zero_division=0)
                if f1 > best_f1:
                    best_f1 = f1
                    best_thresh = thresh
            
            train_acc = ((train_probs > best_thresh) == y_train).mean()
            val_acc = ((val_probs > best_thresh) == y_val).mean()
        
        scheduler.step(val_loss)
        current_lr = optimizer.param_groups[0]['lr']
        
        # Store training history
        training_history.append({
            'epoch': epoch,
            'train_loss': train_loss.item(),
            'val_loss': val_loss.item(),
            'train_acc': train_acc,
            'val_acc': val_acc,
            'best_threshold': best_thresh,
            'val_f1': best_f1,
            'lr': current_lr
        })
        
        # Early stopping based on F1 score improvement
        if best_f1 > best_val_f1:
            best_val_f1 = best_f1
            patience_counter = 0
        else:
            patience_counter += 1
            
        if patience_counter >= config['early_stopping_patience']:
            print(f"Early stopping at epoch {epoch+1} (best F1: {best_val_f1:.3f})")
            break
        
        # Print progress every 10 epochs
        if (epoch + 1) % 10 == 0 or epoch == 0:
            print(f"Epoch {epoch+1:3d}: train_loss={train_loss:.4f}, val_loss={val_loss:.4f}, "
                  f"val_f1={best_f1:.3f}, thresh={best_thresh:.2f}, lr={current_lr:.6f}")
    
    print(f"Training completed after {epoch+1} epochs (best F1: {best_val_f1:.3f})")
    return model, training_history

# Keep the evaluate_model function the same
def evaluate_model(model, X_test, y_test):
    """Evaluate using binary classification metrics with optimal threshold"""
    model.eval()
    X_test_tensor = tc.tensor(X_test, dtype=tc.float32, device=config['device'])
    
    with tc.no_grad():
        logits = model(X_test_tensor).squeeze()
        probs = tc.sigmoid(logits).cpu().numpy()
    
    # Find optimal threshold
    from sklearn.metrics import f1_score
    thresholds = np.arange(0.2, 0.8, 0.05)
    best_f1 = 0
    best_thresh = 0.5
    
    for thresh in thresholds:
        f1 = f1_score(y_test, (probs > thresh).astype(int), zero_division=0)
        if f1 > best_f1:
            best_f1 = f1
            best_thresh = thresh
    
    # Calculate comprehensive metrics
    from sklearn.metrics import roc_auc_score, average_precision_score, accuracy_score, classification_report
    from sklearn.metrics import precision_score, recall_score, balanced_accuracy_score
    
    try:
        # AUC and Average Precision (threshold-independent)
        auc = roc_auc_score(y_test, probs)
        ap = average_precision_score(y_test, probs)
        
        # Metrics with optimal threshold
        predictions = (probs > best_thresh).astype(int)
        accuracy = accuracy_score(y_test, predictions)
        balanced_acc = balanced_accuracy_score(y_test, predictions)
        precision = precision_score(y_test, predictions, zero_division=0)
        recall = recall_score(y_test, predictions, zero_division=0)
        f1 = f1_score(y_test, predictions, zero_division=0)
        
        print(f"  Optimal threshold: {best_thresh:.3f}")
        print(f"  Classification Report (threshold={best_thresh:.3f}):")
        print(classification_report(y_test, predictions, zero_division=0, 
                                  target_names=['Phase < 3', 'Phase 3+']))
        
    except ValueError as e:
        print(f"  Warning: Metric calculation issue: {e}")
        auc = ap = accuracy = balanced_acc = precision = recall = f1 = 0.0
        best_thresh = 0.5
    
    return auc, ap, accuracy, balanced_acc, precision, recall, f1, best_thresh, probs

print("Training functions defined âœ“")
```

```python
def compute_lrp_explanations(model, X_test, test_idx, feature_names):
    """Compute LRP explanations for test samples with GPU memory management"""
    
    model.eval()
    
    # Process in smaller batches to manage GPU memory
    batch_size = 32
    explanations = []
    predictions = []
    
    print(f"Computing LRP explanations for {len(X_test)} test samples...")
    
    for i in range(0, len(X_test), batch_size):
        # Clear cache before each batch
        if tc.cuda.is_available():
            tc.cuda.empty_cache()
            
        batch_end = min(i + batch_size, len(X_test))
        X_batch = tc.tensor(X_test[i:batch_end], dtype=tc.float32, device=config['device'])
        
        # Get predictions for batch
        with tc.no_grad():
            batch_preds = model(X_batch)
            predictions.extend(batch_preds.cpu().numpy().flatten())
        
        # Compute LRP for each sample in batch
        for j, sample_idx in enumerate(range(i, batch_end)):
            # Set iteration for LRP
            for module in model.modules():
                if hasattr(module, 'iteration'):
                    module.iteration = j
            
            sample = X_batch[j:j+1]
            pred = model(sample)
            lrp_scores = model.relprop(pred)
            explanations.append(lrp_scores.cpu().numpy().flatten())
    
    explanations = np.array(explanations)
    
    # Create explanation DataFrame
    explanation_df = pd.DataFrame(explanations, columns=feature_names)
    explanation_df['prediction'] = predictions
    explanation_df['sample_idx'] = test_idx
    
    print(f"âœ… LRP explanations computed successfully!")
    return explanation_df

def analyze_feature_importance(explanation_df, feature_names, top_k=15):
    """Analyze feature importance from LRP explanations with clinical insights"""
    
    # Calculate mean absolute importance for each feature
    feature_importance = explanation_df[feature_names].abs().mean().sort_values(ascending=False)
    
    print(f"\nðŸ”¬ TOP {top_k} MOST PREDICTIVE TRIAL CHARACTERISTICS")
    print("=" * 80)
    
    # Categorize features for better interpretation
    keyword_features = []
    mesh_features = []
    other_features = []
    
    for i, (feature, importance) in enumerate(feature_importance.head(top_k).items()):
        # Categorize feature type
        if feature.startswith('keyword_'):
            category = "ðŸ“‹ STUDY DESIGN"
            clean_name = feature.replace('keyword_', '').replace('_', ' ').title()
            keyword_features.append((clean_name, importance))
        elif feature.startswith('mesh_'):
            category = "ðŸ§¬ DISEASE FOCUS"
            clean_name = feature.replace('mesh_', '').replace('_', ' ').title()
            mesh_features.append((clean_name, importance))
        elif feature.startswith('drug_class_'):
            category = "ðŸ’Š DRUG CLASS"
            clean_name = feature.replace('drug_class_', '').replace('_', ' ').title()
            other_features.append((clean_name, importance))
        else:
            category = "ðŸ¥ TRIAL INFO"
            clean_name = feature.replace('_', ' ').title()
            other_features.append((clean_name, importance))
        
        print(f"{i+1:2d}. {category:<15} {clean_name:<35} {importance:.6f}")
    
    return feature_importance, keyword_features, mesh_features, other_features

def interpret_clinical_insights(keyword_features, mesh_features, other_features):
    """Provide clinical interpretation of the most important features"""
    
    print(f"\nðŸ§  CLINICAL TRIAL SUCCESS INSIGHTS")
    print("=" * 60)
    
    if keyword_features:
        print(f"\nðŸ“‹ STUDY DESIGN FACTORS:")
        print("-" * 30)
        for feature, importance in keyword_features[:5]:
            if 'randomized' in feature.lower():
                print(f"âœ… {feature}: Randomization increases advancement probability")
            elif 'double blind' in feature.lower() or 'blinded' in feature.lower():
                print(f"âœ… {feature}: Blinding improves trial quality and success")
            elif 'efficacy' in feature.lower():
                print(f"ðŸŽ¯ {feature}: Efficacy focus predicts Phase 3+ advancement")
            elif 'placebo' in feature.lower():
                print(f"âš–ï¸  {feature}: Placebo controls indicate rigorous methodology")
            elif 'safety' in feature.lower():
                print(f"ðŸ›¡ï¸  {feature}: Safety focus affects advancement patterns")
            elif 'single arm' in feature.lower():
                print(f"ðŸ“Š {feature}: Single-arm design impacts success rates")
            elif 'multiple arm' in feature.lower():
                print(f"ðŸ”„ {feature}: Multi-arm design influences advancement")
            else:
                print(f"ðŸ“ˆ {feature}: Important design characteristic")
    
    if mesh_features:
        print(f"\nðŸ§¬ DISEASE SCOPE FACTORS:")
        print("-" * 30)
        for feature, importance in mesh_features[:5]:
            if 'als' in feature.lower() or 'amyotrophic' in feature.lower():
                print(f"ðŸŽ¯ {feature}: ALS-specific focus affects success patterns")
            elif 'frontotemporal' in feature.lower():
                print(f"ðŸ§  {feature}: Broader neurodegeneration scope influences outcomes")
            elif 'alzheimer' in feature.lower():
                print(f"ðŸ”— {feature}: Cross-disease approach impacts advancement")
            elif 'parkinson' in feature.lower():
                print(f"ðŸŒ {feature}: Multi-indication strategy affects success")
            else:
                print(f"ðŸ”¬ {feature}: Disease classification impacts advancement")
    
    if other_features:
        print(f"\nðŸ¥ OTHER TRIAL FACTORS:")
        print("-" * 30)
        for feature, importance in other_features[:3]:
            print(f"ðŸ“Š {feature}: Trial characteristic influences success")

def generate_actionable_recommendations(feature_importance, top_k=10):
    """Generate actionable recommendations based on feature importance"""
    
    print(f"\nðŸŽ¯ ACTIONABLE RECOMMENDATIONS FOR TRIAL DESIGN")
    print("=" * 60)
    
    top_features = feature_importance.head(top_k)
    
    design_recommendations = []
    disease_recommendations = []
    
    for feature, importance in top_features.items():
        if feature.startswith('keyword_'):
            keyword = feature.replace('keyword_', '').replace('_', ' ')
            if 'randomized' in keyword:
                design_recommendations.append("âœ… Prioritize randomized trial designs")
            elif 'double blind' in keyword or 'blinded' in keyword:
                design_recommendations.append("âœ… Implement blinding when feasible")
            elif 'efficacy' in keyword:
                design_recommendations.append("ðŸŽ¯ Focus on efficacy endpoints over safety-only")
            elif 'placebo control' in keyword:
                design_recommendations.append("âš–ï¸  Include placebo controls for rigor")
        elif feature.startswith('mesh_'):
            disease = feature.replace('mesh_', '').replace('_', ' ')
            if 'frontotemporal' in disease:
                disease_recommendations.append("ðŸ§  Consider broader neurodegeneration scope")
            elif 'alzheimer' in disease:
                disease_recommendations.append("ðŸ”— Explore cross-disease mechanisms")
    
    if design_recommendations:
        print(f"\nðŸ“‹ STUDY DESIGN RECOMMENDATIONS:")
        for rec in design_recommendations[:5]:
            print(f"  {rec}")
    
    if disease_recommendations:
        print(f"\nðŸ§¬ DISEASE FOCUS RECOMMENDATIONS:")
        for rec in disease_recommendations[:3]:
            print(f"  {rec}")
    
    print(f"\nðŸ’¡ STRATEGIC INSIGHTS:")
    print(f"  â€¢ Trials with rigorous methodology (randomized, blinded) advance more often")
    print(f"  â€¢ Efficacy-focused trials outperform safety-only studies")
    print(f"  â€¢ Disease scope and indication strategy impact success rates")
    print(f"  â€¢ Study design choices are measurably predictive of advancement")

print("Enhanced LRP analysis functions defined âœ…")
```
